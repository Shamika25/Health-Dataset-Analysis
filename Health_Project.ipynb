## Question 2: Read the Male and Female Dataset

### Purpose

The purpose of this step is to read the male and female NHANES datasets and store them as NumPy matrices for further analysis.

```python
# Import NumPy
import numpy as np

# Read the datasets as NumPy matrices

male = np.loadtxt(
    "nhanes_adult_male_bmx_2020.csv",
    delimiter=",",
    skiprows=19
)

female = np.loadtxt(
    "nhanes_adult_female_bmx_2020.csv",
    delimiter=",",
    skiprows=19
)

# Display shape of both matrices

print("Male Matrix Shape:", male.shape)
print("Female Matrix Shape:", female.shape)
```

# Question 3: Plot Histograms of Female and Male Weights

## Purpose

The purpose of this step is to compare the weight distribution of female and male participants using two histograms. Both histograms are displayed in a single figure using subplots with the same x-axis limits for easy comparison.

```python
# Extract weight column
female_weights = female[:, 0]
male_weights = male[:, 0]

# Create one figure
plt.figure(figsize=(8,8))

# Top subplot - Female
plt.subplot(2,1,1)
plt.hist(female_weights, bins=20, edgecolor="black")
plt.title("Female Weight Distribution")
plt.xlabel("Weight (kg)")
plt.ylabel("Frequency")
plt.xlim(20,180)

# Bottom subplot - Male
plt.subplot(2,1,2)
plt.hist(male_weights, bins=20, edgecolor="black")
plt.title("Male Weight Distribution")
plt.xlabel("Weight (kg)")
plt.ylabel("Frequency")
plt.xlim(20,180)

plt.tight_layout()
plt.show()
```

The histograms show the distribution of body weight for female and male participants. Most participants have weights around the middle range. Male participants generally have higher body weights than female participants. Using the same x-axis limits makes it easier to compare the two distributions.

# Question 5: Compute Basic Numerical Aggregates of Male and Female Weights

## Purpose

The purpose of this step is to calculate the basic statistical measures of male and female weights. These measures include location (mean, median), dispersion (standard deviation, variance, range), and shape (skewness and kurtosis). These statistics help compare the weight distributions of male and female participants.

```python
# Import required library

from scipy import stats

# Extract weight column

male_weights = male[:,0]
female_weights = female[:,0]

# Male statistics

print("----- Male Weight Statistics -----")
print("Mean:", np.mean(male_weights))
print("Median:", np.median(male_weights))
print("Standard Deviation:", np.std(male_weights))
print("Variance:", np.var(male_weights))
print("Minimum:", np.min(male_weights))
print("Maximum:", np.max(male_weights))
print("Range:", np.ptp(male_weights))
print("Skewness:", stats.skew(male_weights))
print("Kurtosis:", stats.kurtosis(male_weights))

# Female statistics

print("\n----- Female Weight Statistics -----")
print("Mean:", np.mean(female_weights))
print("Median:", np.median(female_weights))
print("Standard Deviation:", np.std(female_weights))
print("Variance:", np.var(female_weights))
print("Minimum:", np.min(female_weights))
print("Maximum:", np.max(female_weights))
print("Range:", np.ptp(female_weights))
print("Skewness:", stats.skew(female_weights))
print("Kurtosis:", stats.kurtosis(female_weights))
```

## Interpretation

The male and female weight distributions were compared using measures of location, dispersion, and shape.

The mean and median indicate the central tendency of the data. The group with the higher mean has a higher average body weight.

The standard deviation, variance, and range indicate the spread of the data. The distribution with the higher standard deviation has greater dispersion, meaning the weights vary more among the participants.

The skewness value indicates the symmetry of the distribution. A positive skewness value means the distribution is right-skewed, while a negative skewness value means it is left-skewed. A skewness value close to zero indicates that the distribution is approximately symmetric.

The kurtosis value describes the shape of the distribution. A higher kurtosis value indicates heavier tails and more extreme values, whereas a lower kurtosis value indicates a flatter distribution.

Overall, these statistical measures help compare the average weight, variability, and distribution shape of male and female participants.

# Question 6: Add BMI as the Eighth Column

## Purpose

The purpose of this step is to calculate the Body Mass Index (BMI) of each female participant and add it as the eighth column of the female matrix.

```python
# Calculate BMI

female_bmi = female[:,0] / ((female[:,1]/100)**2)

# Add BMI as 8th column

female = np.column_stack((female, female_bmi))

# Display first five rows

print(female[:5])
```

## Interpretation

The BMI values were calculated using weight and height measurements and added as the eighth column of the female matrix. This new column provides an additional health indicator that will be used in further analysis.

# Question 7: Standardize the Female Dataset

## Purpose

The purpose of this step is to standardize all columns of the female dataset using z-scores. Standardization converts all variables to the same scale with a mean of 0 and a standard deviation of 1.

```python
# Standardize the female matrix

zfemale = stats.zscore(female)

# Display first five rows

print(zfemale[:5])
```

## Interpretation

The standardized female matrix (zfemale) was created successfully. Each value now represents how many standard deviations it is above or below the mean. Standardization allows different body measurements to be compared on the same scale.

# Question 9: Compute Waist-to-Height Ratio and Waist-to-Hip Ratio

## Purpose

The purpose of this step is to calculate Waist-to-Height Ratio (WHtR) and Waist-to-Hip Ratio (WHR) for both male and female participants. These ratios are useful health indicators for assessing obesity and cardiovascular risk.

```python
# Calculate Waist-to-Height Ratio (WHtR)

male_whtr = male[:,6] / male[:,1]
female_whtr = female[:,6] / female[:,1]

# Calculate Waist-to-Hip Ratio (WHR)

male_whr = male[:,6] / male[:,5]
female_whr = female[:,6] / female[:,5]

# Add the two new columns to the male matrix

male = np.column_stack((male, male_whtr, male_whr))

# Add the two new columns to the female matrix

female = np.column_stack((female, female_whtr, female_whr))

# Display first five rows

print("Male Matrix (First 5 Rows)")
print(male[:5])

print("\nFemale Matrix (First 5 Rows)")
print(female[:5])
```

## Interpretation

The Waist-to-Height Ratio (WHtR) and Waist-to-Hip Ratio (WHR) were successfully calculated for both male and female participants. These two ratios were added as new columns to the respective matrices.

The Waist-to-Height Ratio indicates whether a person's waist circumference is appropriate for their height, while the Waist-to-Hip Ratio shows how body fat is distributed around the waist and hips. These measures provide additional health information beyond BMI and are useful for assessing the risk of obesity and cardiovascular diseases.

# Question 11: Advantages and Disadvantages of BMI, Waist-to-Height Ratio, and Waist-to-Hip Ratio

## Purpose

The purpose of this step is to understand the strengths and limitations of three common health indicators: Body Mass Index (BMI), Waist-to-Height Ratio (WHtR), and Waist-to-Hip Ratio (WHR).

## Advantages and Disadvantages

### 1. Body Mass Index (BMI)

**Advantages**
- Easy to calculate using height and weight.
- Commonly used to classify underweight, normal weight, overweight, and obesity.
- Useful for studying large populations.

**Disadvantages**
- Does not distinguish between muscle and fat.
- Does not show where body fat is stored.
- May incorrectly classify athletes or muscular people as overweight.

---

### 2. Waist-to-Height Ratio (WHtR)

**Advantages**
- Better indicator of abdominal fat.
- Helps predict the risk of heart disease and diabetes.
- Easy to calculate and understand.

**Disadvantages**
- Requires accurate waist measurement.
- May not be equally suitable for all age groups.

---

### 3. Waist-to-Hip Ratio (WHR)

**Advantages**
- Measures body fat distribution.
- Useful for identifying abdominal obesity.
- Helps assess cardiovascular health risk.

**Disadvantages**
- Hip and waist measurements may vary depending on the measurement technique.
- Less useful for monitoring overall body weight changes.

# Question 12: Display Standardized Measurements for the Lowest and Highest BMI Participants

## Purpose

The purpose of this step is to identify the five participants with the lowest BMI and the five participants with the highest BMI using the standardized female dataset (zfemale). This helps compare body measurements between individuals with very low and very high BMI values.

```python
# Find BMI column (8th column, index 7)

bmi = female[:,7]

# Sort BMI values in ascending order

sorted_index = np.argsort(bmi)

# Get the indices of the lowest 5 BMI participants

lowest_bmi = sorted_index[:5]

# Get the indices of the highest 5 BMI participants

highest_bmi = sorted_index[-5:]

# Combine the indices

selected = np.concatenate((lowest_bmi, highest_bmi))

# Print standardized body measurements

print("Standardized Body Measurements of Selected Participants")

print(zfemale[selected])
```

## Interpretation

The standardized body measurements of the five participants with the lowest BMI and the five participants with the highest BMI were displayed successfully.

Participants with the lowest BMI generally have lower standardized values for weight, waist circumference, hip circumference, and BMI. In contrast, participants with the highest BMI generally have higher standardized values for these measurements.

This comparison shows that BMI is strongly associated with body weight, waist circumference, and hip circumference. Standardized values make it easier to compare participants because all body measurements are converted to the same scale.
